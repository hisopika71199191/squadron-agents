```python
#Agent Skills - no need to import CodingTools separately
from squadron import Agent, SkillsManager, LATSReasoner
from dotenv import load_dotenv
from squadron.llm import OpenAICompatibleProvider
import os

load_dotenv()

llm = OpenAICompatibleProvider(
    api_key=os.getenv("ALI_API_KEY"),
    base_url=os.getenv("ALI_BASE_URL"),
    model=os.getenv("LLM_MODEL", "qwen3.5-plus"),
    temperature=float(os.getenv("LLM_TEMPERATURE", "0.7")),
    # Use at least 8192 tokens so the JSON response is not truncated mid-array.
    # The skill instructions + prompt can exceed 4096 tokens on their own.
    max_tokens=int(os.getenv("LLM_MAX_TOKENS", "8192")),
)

# tools_workspace = the directory where the agent runs commands and writes files.
# The output .pptx will be created here (e.g. examples/Presentation.pptx).
TOOLS_WORKSPACE = "/Users/dk_studio/Documents/project/squadron-agents/examples"

# Discover the pptx skill from src/squadron/skills/pptx/SKILL.md
skills = SkillsManager(
    workspace_path="/Users/dk_studio/Documents/project/squadron-agents/src/squadron",
    skills_dirs=["skills"],   # maps to src/squadron/skills/
    tools_workspace=TOOLS_WORKSPACE,
)
await skills.discover_skills()  # discovers src/squadron/skills/pptx/SKILL.md

# Create reasoner with skills support
reasoner = LATSReasoner(skills_manager=skills, llm=llm)

# Pass skills directly as tools= — SkillsManager.get_tools() automatically
# provides run_command, write_file, read_file, etc. scoped to tools_workspace.
# The generated .pptx file will be saved inside TOOLS_WORKSPACE.
agent = Agent(name="developer", reasoner=reasoner, llm=llm, tools=skills)

result = await agent.run(
    "Create a PowerPoint presentation of 3 slides to recommend top 3 online games in 2026"
)
import os
pptx_files = [f for f in os.listdir(TOOLS_WORKSPACE) if f.endswith(".pptx")]
print(f"Output files in {TOOLS_WORKSPACE}: {pptx_files}")
```
